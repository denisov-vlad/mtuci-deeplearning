*Распознавание рукописных цифр из набора данных MNIST на Keras*


Использована простая нейронная сеть, состоящая из двух слоев: входного и выходного. В качестве функций активации используются ReLU и SoftMax. В качестве метода оптимизации используется стохастический градиентный спуск.

![Network](../images/lab1_1.png)


Импорт библиотек (keras, numpy)

```python
from __future__ import print_function
import numpy as np
np.random.seed(1337)  # для повторяемости

from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Convolution2D, MaxPooling2D
from keras.utils import np_utils
from keras import backend as K
```

Исходные параметры

```python
batch_size = 128
nb_classes = 10
nb_epoch = 12

# размер входных изображений
img_rows, img_cols = 28, 28
# число сверточных слоев
nb_filters = 32
# размер зоны для пулинг
pool_size = (2, 2)
# размерность ядра
kernel_size = (3, 3)
```

```python
# Загрузка данных в обучающую и тестовую выборки
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# Reshape матриц в зависимости от бекенда
if K.image_dim_ordering() == 'th':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
print('X_train shape:', X_train.shape)
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
# преобразование матриц в бинарный формат
Y_train = np_utils.to_categorical(y_train, nb_classes)
Y_test = np_utils.to_categorical(y_test, nb_classes)
```

```python
# создание модели
model = Sequential()

model.add(Convolution2D(nb_filters, kernel_size[0], kernel_size[1],
                        border_mode='valid',
                        input_shape=input_shape))
model.add(Activation('relu'))
model.add(Convolution2D(nb_filters, kernel_size[0], kernel_size[1]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=pool_size))
model.add(Dropout(0.5))

model.add(Flatten())
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(nb_classes))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adadelta',
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, batch_size=batch_size, nb_epoch=nb_epoch,
          verbose=1, validation_data=(X_test, Y_test))
```

```python
score = model.evaluate(X_test, Y_test, verbose=1)

print('Test score:', score[0])
print('Test accuracy:', score[1])
```

```python
model.save('lab1_model.h5')
```
